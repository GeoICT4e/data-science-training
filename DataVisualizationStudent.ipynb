```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('data\\CarPrice.csv')

cars
```

```python
cars.head()
```

```python
#cars = cars.drop('car_ID',axis=1)


cars[["symboling" ,'wheelbase']] = cars[['symboling', 'wheelbase']].astype(float)
```

```python
cars
```

```python
cars.duplicated().sum()
```

```python
cars['CarName'].value_counts()
```

```python
ax1=cars['CarName'].value_counts()[:20].plot(kind='bar',stacked=False, colormap = 'Set2')
ax1.title.set_text('Car name 1111')
for c in ax1.containers:
    ax1.bar_label(c, c='g')
plt.ylabel("Count of Cars",fontweight = 'bold')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plt.subplot(3,3,1)
sns.boxplot(x = 'doornumber', y = 'price', data=cars)
plt.subplot(3,3,2)
sns.boxplot(x = 'fueltype', y = 'price', data=cars)
plt.subplot(3,3,3)
sns.boxplot(x = 'aspiration', y = 'price', data=cars)
plt.subplot(3,3,4)
sns.boxplot(x = 'carbody', y = 'price', data=cars)
plt.subplot(3,3,5)
sns.boxplot(x = 'enginelocation', y = 'price', data=cars)
plt.subplot(3,3,6)
sns.boxplot(x = 'drivewheel', y = 'price', data=cars)
plt.subplot(3,3,7)
sns.boxplot(x = 'enginetype', y = 'price', data=cars)
plt.subplot(3,3,8)
sns.boxplot(x = 'cylindernumber', y = 'price', data=cars)
plt.subplot(3,3,9)
sns.boxplot(x = 'fuelsystem', y = 'price', data=cars)
plt.show()
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(1,3,1)
plt1 = cars['cylindernumber'].value_counts().plot(kind='bar', colormap = 'Set2')
for container in plt1.containers:
    plt1.bar_label(container, c='g')
plt.title('Number of cylinders')
plt1.set(xlabel = 'Number of cylinders', ylabel='Frequency of Number of cylinders')

plt.subplot(1,3,2)
plt1 = cars['fueltype'].value_counts().plot(kind='bar',  colormap = 'Set2')
for container in plt1.containers:
    plt1.bar_label(container, c='g')
plt.title('Fuel Type')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of Fuel type')

plt.subplot(1,3,3)
plt1 = cars['carbody'].value_counts().plot(kind='bar',  colormap = 'Set2')
for container in plt1.containers:
    plt1.bar_label(container, c='g')
plt.title('Car body')
plt1.set(xlabel = 'Car Body', ylabel='Frequency of Car Body')
plt.show()
```
